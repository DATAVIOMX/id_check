```python
from mtcnn.mtcnn import MTCNN
from PIL import Image
import pytesseract
import cv2
import os
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread("ifes/5_front.jpg")
plt.imshow(image)
plt.show()
```

```python
def detect_ife(path):
    detector = MTCNN()
    img = cv2.imread(path)
    result = detector.detect_faces(img)
    text = pytesseract.image_to_string(Image.open(path), 'spa')
    if result[0]["confidence"] >= 0.5 and "INSTITUTO" in text:
        return "IFE DETECTED"
    else:
        return "IFE NOT DETECTED"
```

```python
detect_ife("ifes/5_front.jpg")
```
